### Liquidity Ratio
#### current_ratio= current_assets / current_liabilities <br>
#### acidtest_ratio = current_assets – inventories / current_liabilities <br>
#### cash_ratio = cash_Cash_equivalents / current_liabilities <br>
operating_cash_flow ratio = operating_cash_flow / current_liabilities <br>
<br>
liquidity_ratio=[current_ratio, acid-test_ratio, cash_ratio, operating_cash_flow_ratio]

current_asset= cash, cash equivalents, accounts receivable, stock inventory, marketable securities, pre-paid liabilities, and other liquid assets.

### Leverage Financial Ratios
#### Debt ratio = Total liabilities / Total assets <br>
#### Debt to equity ratio = Total liabilities / Shareholder’s equity <br>
#### Interest coverage ratio = Operating income / Interest expenses <br>
Debt service coverage ratio = Operating income / Total debt service

### Efficiency Ratios

#### Asset turnover ratio = Total Revenue / Average total assets <br>
#### Inventory turnover ratio = Cost of goods sold / Average inventory <br>
Receivables turnover ratio = Net credit sales / Average accounts receivable <br>
Days sales in inventory ratio = 365 days / Inventory turnover ratio <br>

### Profitabilities Ratios

#### Gross margin ratio = Gross profit / Total Revenue <br>
#### Operating margin ratio = Operating income / Net sales <br>
#### Return on assets ratio = Net income / Total assets <br>
#### Return on equity ratio = Net income / Shareholder’s equity

### Market Value Ratios

Book value per share ratio = (Shareholder’s equity – Preferred equity) / Total common shares outstanding <br>
Dividend yield ratio = Dividend per share / Share price <br>
Earnings per share ratio = Net earnings / Total shares outstanding <br>
Price-earnings ratio = Share price / Earnings per share
